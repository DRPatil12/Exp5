```python
#changing/adding key:value in dictionary
#Method 1:
car={"brand": "Ford", "model": "Mustang", "year": 1964}
car.update({"color": "White"})
print("Method 1:",car)
#Method 2:
car1={"brand": "Ford", "model": "Mustang", "year": 1964}
car1["color"]="White"
print("Method 2:",car1)
```

```python
#To print only values/keys:
car={"brand": "Ford", "model": "Mustang", "year": 1964}
print(car.values())
print(car.keys())
```

```python
#retrival of values:
student= {
    "roll": 101,
    "name": "Dhruv",
    "branch": "EnTC",
}
print(student["roll"])
```

```python
#Removing branch from dictionary using .pop
student= {
    "roll": 101,
    "name": "Dhruv",
    "branch": "EnTC",
}
student.pop("branch")
print(student)
```

```python
'''Problem 2: A dictionary stores student names and marks.
Write a python program to search for a students marks using the name entered by
the user.
If the student is not found, display a suitable message.'''
marks={"Dhruv": 20, "Ronit": 17, "Rishabh":18}
name= input("Enter the name of student:")
print(marks.get(name, "Student not found in database"))
```

```python
'''Problem 3: A dictionary stores usernames and passwords
Write a Python program to validate user login.'''
users={"admin": "1234", "student": "abcd"}
username= input("Enter the username:")
password= input("Enter the password:")
if username in users:
    users.get("Username")==password
    print("Login Successful")
else:
    print("Login Failed")
```

```python
'''Problem 4: A dictionary stores studeny names and marks.
Write a python program to find the student who scored the highest marks.'''
marks={"a": 1, "b": 2, "c": 3}
topper = max(marks, key=marks.get)
print("topper:",topper)
print("marks:", marks[topper])
```
